```python
#W4A8
def reverse(x):
    nw = 0;
    while (x > 0):
        nw = nw * 10 + (x % 10)
        x //= 10;
    return nw

def palindrome(x):
    if (x == reverse(x)): return True
    return False

n = int(input())
cnt = 0;
while (palindrome(n) == False):
    cnt += 1;
    n = n + reverse(n)
print (cnt, n)
```

```python
#W4A9
def distinct(x):
    a = list(range(10))
    for i in range(10):
        a[i] = False
    while (x > 0):
        if (a[x % 10] == True): return False;
        a[x % 10] = True
        x //= 10
    return True

n = int(input())
flag = False
for i in range(1, n + 1):
    x = int(i**0.5)
    if ((x * x == i) and (distinct(i) == True)):
        flag = True
        print(i, end = ' ')
if (flag == False):
    print("no number")
```

```python
#W4A13
def sum_div(x):
    sum = 1
    for i in range(2, int(x**0.5) + 1):
        if (x % i == 0):
            sum += i
            if (i != x // i): sum += x // i
    return sum

a, b = map(int, input().split())
if (sum_div(a) == b and sum_div(b) == a): print("true")
else: print("false")
```

```python
#W4A14
m, n = map(int, input().split())
ans = 0
for i in range(1, int(m**0.5) + 1):
    if (m % i == 0):
        if (n % i == 0): ans = max(ans, i)
        if (i != m // i):
            if (n % (m // i) == 0): ans = max(ans, m // i)
print(ans)
```
