```python
#W5A1
def Max_Of_Two(a, b):
    if (a > b): return a
    else: return b

if __name__ == '__main__':
    a, b = map(int, input().split())
    print(Max_Of_Two(a, b))
```

```python
#W5A2
def Swap(a, b):
    c = a
    a = b
    b = c
    return (a, b)

if __name__ == '__main__':
    a, b = map(int, input().split())
    print(Swap(a, b))
```

```python
#W5A3
import math as mth

def Prime_Num(x):
    if (x <= 1): return False
    for i in range(2, int(mth.sqrt(x) + 1)):
        if (x % i == 0): return False
    return True

if __name__ == '__main__':
    n = int(input())
    if (Prime_Num(n)): print("Yes")
    else: print("No")
```

```python
#W5A4
import math as mth

def Sum_Div(x):
    sum = 0
    for i in range(1, int(mth.sqrt(x)) + 1):
        if (x % i == 0):
            sum += i
            if (i != x / i):
                sum += x / i
    return sum

if __name__ == '__main__':
    n = int(input())
    x = Sum_Div(n) - n
    if (n == x): print("Yes")
    else: print("No")
```

```python
#W5A5
def solve(a, k):
    p = 0
    for i in range(len(a)):
        if (a[i] == k):
            p = i + 1
    return p

if __name__ == '__main__':
    k = int(input())
    a = list(map(int, input().split()))
    print(solve(a, k))
```

```python
#W5A6
def fac(n):
    ans = 1
    for i in range(1, n + 1):
        ans *= i
    return ans

if __name__ == '__main__':
    n = int(input())
    print(fac(n))
```

```python
#W5A8
def Hamming(a, b):
    c = a ^ b
    return(c.bit_count())

if __name__ == '__main__':
    a, b = map(int, input().split())
    print(Hamming(a, b))
```

```python
#W5A9
def solve(n):
    ans = 0
    while (n > 0):
        ans += (n % 10)
        n //= 10
    return ans

if __name__ == '__main__':
    n = int(input())
    print(solve(n))
```

```python
#W5A10

def check(s, t):
    if (len(s) != len(t)):
        return False
    d1 = [0] * 30
    d2 = [0] * 30
    for i in range(len(s)):
        if ((d1[ord(s[i]) - ord('a') + 1] == 0) and (d2[ord(t[i]) - ord('a') + 1] == 0)):
            d1[ord(s[i]) - ord('a') + 1] = ord(t[i]) - ord('a') + 1
            d2[ord(t[i]) - ord('a') + 1] = ord(s[i]) - ord('a') + 1
        else:
            if ((d1[ord(s[i]) - ord('a') + 1] != ord(t[i]) - ord('a') + 1) or (d2[ord(t[i]) - ord('a') + 1] != ord(s[i]) - ord('a') + 1)):
                return False
    return True

if __name__ == '__main__':
    s, t = map(str, input().split())
    print(check(s, t))
```
